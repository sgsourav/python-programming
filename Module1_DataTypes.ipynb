## Basic Data Types

Fundamental data types in Python are `int` (Integers), `float` (Decimal Numbers), `str` (Strings), `bool` (Boolean), and `complex` (Complex Numbers).

```python
x = 3
print(x, "is of type", type(x))
```

```python
x = 3.0
print(x, "is of type", type(x))
```

```python
x = 3e10
print(x, "is of type", type(x))
```

```python
x = "3"
print(x, "is of type", type(x))
```

```python
x = True
print(x, "is of type", type(x))
```

```python
x = "True"
print(x, "is of type", type(x))
```

```python
x = 1 + 2j
print(x, "is of type", type(x))
```

```python
x = complex(1,2)
print(x, "is of type", type(x))
```

## Basic Operations

The fundamental data types support basic arithmetic and logical operations in Python, as is expected from any standard computing environment.

```python
# Addition
5 + 2
```

```python
# Subtraction 
5 - 2
```

```python
# Multiplication
5 * 2
```

```python
# Exact Division
5 / 2
```

```python
# Quotient
5 // 2
```

```python
# Remainder
5 % 2 
```

```python
# Exponentiation
5 ** 2
```

```python
# Assignment
5 = 2
```

```python
# Equal to
5 == 2
```

```python
# Not Equal to
5 != 2
```

```python
# Less than
5 < 2
```

```python
# Greater than
5 > 2
```

```python
# Less than or Equal to
5 <= 2
```

```python
# Greater than or Equal to
5 >= 2
```

```python
# Logical OR
True or False
```

```python
# Logical AND
True and False
```

```python
# Combined Logic
(5 != 2) or (5 < 2)
```

```python
# Combined Logic
(5 >= 2) and (not (5 == 2))
```

```python
# Combined Inequality
5 > 2 > 1
```

```python
# Combined Inequality
5 > 2 < 10
```

```python
# Combined Logic -- Compute and Print
print( not (True and False), "==", not True or not False)
```

```python
# Bitwise Logic
a, b = 5, 2
print('a =', a, "=", bin(a))
print('b =', b, "=", bin(b))
print()   # print blank line

print('a & b =', a & b, "=", bin(a & b))
print('a | b =', a | b, "=", bin(a | b))
print('a ^ b =', a ^ b, "=", bin(a ^ b))
```

#### Quick Tasks

- What is the value of `(5 != 2) or (5/0 < 2)`?    
- What is the value of `(5 != 2) and (5/0 < 2)`?

## Data Type Conversion

One may easily convert between the fundamental data types -- `int` and `str` -- especially during base shifts (binary, hex, etc).

```python
x = bin(125)
print(x, "is of type", type(x))
```

```python
x = hex(125)
print(x, "is of type", type(x))
```

```python
x = 0b1101
print(x, "is of type", type(x))
```

```python
x = 0x1a
print(x, "is of type", type(x))
```

```python
x = int("11")
print(x, "is of type", type(x))
```

```python
x = int("11", 2)
print(x, "is of type", type(x))
```

```python
x = int("11", 16)
print(x, "is of type", type(x))
```

```python
x = str(11)
print(x, "is of type", type(x))
```

```python
x = str(0b11)
print(x, "is of type", type(x))
```

#### Quick Tasks

- Add two binary numbers `1101`, `1001` and print the result in binary.    
- Add two hexadecimal numbers `2A3F`, `1D21` and print the result in hex.
